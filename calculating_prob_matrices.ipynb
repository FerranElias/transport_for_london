```python
import pandas as pd

#paths in my computer
datapath = 'Y:\\okofm\\tfl\\journey_data\\'

#paths at the university cluster
#datapath = 'Y:\\tfl\\journey_data\\data_random_sample_2012_2014.csv'

data = pd.read_csv(datapath+'data_random_sample_2012_2014.csv')

data

new_dates = []

for index, row in data.iterrows():
    if len(str(row['Start Date']))==19:
        new_dates.append(row['Start Date'][:-3])
    elif len(str(row['Start Date']))==16:
        new_dates.append(row['Start Date'])
    elif len(str(row['Start Date']))==3:
        new_dates.append(row['Start Date'])
        
        #print(row['Start Date'][:-3])

#new_dates
se = pd.Series(new_dates)
data['Start Date'] = se.values
data

data.dropna(axis=0, subset=['Start Date'], inplace=True)

#transform duration into minutes
data['Duration'] = data['Duration']/60

#adding day of the week, year, quarter, month, week, day, hour, minute
data['year'] = pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.year
data['quarter'] = pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.quarter
data['month'] = pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.month
data['week'] =  pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.week
data['day'] =  pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.day
data['hour'] =  pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.hour
data['minute'] =  pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.minute
data['day_of_week'] = pd.to_datetime(data['Start Date'],format='%d/%m/%Y %H:%M').dt.dayofweek
days = {0:'Mon',1:'Tues',2:'Weds',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
data['day_of_week'] = data['day_of_week'].apply(lambda x: days[x])

data
```

```python
#checking number of stations matches
print(len(data['StartStation Id'].unique()))
print(len(data['EndStation Id'].unique()))
```

```python
#getting number of journeys by start and end station
journey_counts = data.groupby(['StartStation Id','EndStation Id']).count()
columns_to_drop = journey_counts.columns
journey_counts['Number of Journeys'] = journey_counts['Unnamed: 0']
journey_counts.drop(columns_to_drop, axis=1, inplace=True)
journey_counts
```

```python
#number of total journeys starting at each station
journey_total = journey_counts.groupby(level=0).sum()
journey_total
```

```python
#getting the probabilities
journey_prob = journey_counts.merge(journey_total, how='inner', left_index='StartStation Id', right_index='StartStation Id')
journey_prob['Probability'] = journey_prob['Number of Journeys_x']/journey_prob['Number of Journeys_y']
journey_prob
```

```python
#checking they add to 1
journey_prob['Probability'].groupby(level=0).sum()
```

```python
journey_prob = journey_prob['Probability']
journey_prob
```

```python
prob_matrix = journey_prob.unstack()
```

```python
prob_matrix.fillna(0, inplace=True)
prob_matrix
```
